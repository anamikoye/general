```python
#Initializations 
vocab = []
coefficients = []
reviews = []
featurespace = []

#Fill in the vocabulary list from logreg document 
with open('logreg_coefficients.txt') as f:
    data = f.readlines()
    for line in data:
        split_data = line.split()      
        if len(split_data) ==2: #A word and a coefficient
            vocab.append(split_data[0])
            coefficients.append(split_data[1])
            
        
#Fill in reviews array from test_movies files
with open('test_movies.txt') as f:
    data = f.readlines()
    for line in data:
        reviews.append(line)

#Fill in the featurespace. A matrix where the rows are movie reviews and columns are words in the vocabulary.
for review in reviews:
    feature_vector = np.zeros(len(vocab))#Create an array of zeros. Add 1 for the bias feature
    for word in review.split():
        try:
            feature_vector[vocab.index(word)]=1
        except:
            #If the word is not in the vocabulary assume it's coefficient to be 0
            pass
    #feature_vector[-1]=1 #Make the last element(bias feature a 1)
    featurespace.append(feature_vector)


#Get the dot product of the feature space and coefficients
dot_product = np.dot(np.array(featurespace,dtype=float),np.array(coefficients,dtype=float))

log_reg = np.exp(dot_product) / (1+np.exp(dot_product)) # P(y = 1 | x, β)

y_hat = [1 if x >=0.5  else 0 for x in log_reg] # y^
print('\n\nFor each of the ten movies, the probability P(y = 1 | x, β) is:\n\n',log_reg ,'\n\nthe prediction yˆ is:\n', y_hat)
```
